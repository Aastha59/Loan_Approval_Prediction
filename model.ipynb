```python
!pip install scikit-learn
from sklearn.model_selection import train_test_split
```

```python
input_data = data.drop(columns=['loan_status'])
output_data = data['loan_status']
```

```python
input_data
```

```python
output_data
```

```python
categorical_cols = input_data.select_dtypes(include=['object']).columns
numeric_cols = input_data.select_dtypes(exclude=['object']).columns
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

encoder = OneHotEncoder(drop='first', sparse_output=False)
input_data_encoded = pd.DataFrame(encoder.fit_transform(input_data[categorical_cols]))
input_data_encoded.columns = encoder.get_feature_names_out(categorical_cols)

# Combine with numeric columns
input_data_final = pd.concat([input_data[numeric_cols].reset_index(drop=True), input_data_encoded], axis=1)


# categorical_cols = ['profession', 'city', 'state']  # match exactly what you see in x_train.columns
# categorical_data = x_train[categorical_cols].astype(str)
# #
# # ✅ Fit OneHotEncoder on categorical data
# encoder = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')
# encoded_cats = encoder.fit_transform(categorical_data)

```

```python
x_train, x_test, y_train, y_test = train_test_split(input_data_final, output_data, test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
!pip install xgboost --upgrade
from xgboost import XGBClassifier

model = XGBClassifier(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)
model.fit(x_train_scaled, y_train)
```

```python
model.score(x_test_scaled, y_test)
```
